```python
# Creates non overlapping pairs of dates, of length duration(days)
# Automatically covers 1 month after the start date
duration = 3
args['dates_train'], args['dates_val'], args['dates_test'] = [], [], []

for dataset, str_date in [('dates_train','Jan 1 2016'), ('dates_train','Jan 1 2017'), 
                          ('dates_val','Jan 1 2018'), ('dates_test','Jan 1 2019')]:

    start = datetime.strptime(str_date, '%b %d %Y')
    args[dataset] = args[dataset] + [(start + timedelta(days=t*duration), start+timedelta(days=(t+1)*duration)) \
                                     for t in range(int(365/duration))]
```

```python
# Load and Format Location and Feature Dataframes
loc = pd.read_hdf(args['loc_data'], key='loc')
    # Datetime: Date
    
df = pd.read_csv(args['real_data'], sep='|')
df = df.loc[df.AdmitDate<df.DischargeDate,:]
df['DischargeDate'] = pd.to_datetime(df.DischargeDate)
df['AdmitDate'] = pd.to_datetime(df.AdmitDate)
df['Date'] = pd.to_datetime(df.Date)
df['CollectionDateCdiff'] = pd.to_datetime(df.CollectionDateCdiff)
    # Datetime: AdmitDate, DischargeDate, Date, CollectionDateCdiff
```

```python
def df_to_network(loc, df, args):

    ## INCLUSION/EXCLUSION, LABELS ##
    # Identify last_day of previous dataset
    try:
        args['dates_val'].index((args['start_date'],args['end_date']))
        last_day = max([y for (x,y) in args['dates_train']])
    except: 
        try:
            args['dates_test'].index((args['start_date'],args['end_date']))
            last_day = max([y for (x,y) in args['dates_val']])
        except:
            last_day = None
            
    # Keep within date range [start, end)
    loc = loc.loc[(loc.Date>=args['start_date'])&(loc.Date<args['end_date']),:].copy()
    df = df.loc[(df.Date>=args['start_date'])&(df.Date<args['end_date']),:].copy()
    args['T'] = len(loc.Date.unique())
    print('Start Date {}, End Date {}, T {}'.format(args['start_date'],args['end_date'],args['T']))
    
    # Network Inclusion Criteria wrt start/end date
    # 1. Positive prior to end date.
    # 2. Prior case of CDI.
    # 3. LOS of 7 prior to end date.
    df['nic_Cdiff'] = 1*((df.Cdiff==1) & (df.CollectionDateCdiff <args['end_date']))    
    df['nic_LOS'] = 1*((args['end_date']-df.location
                        dmitDate.astype('<M8[D]')).dt.days>6)
    
    df = df.loc[(df.nic_LOS==1) | (df.nic_Cdiff==1), :]
    loc = loc.merge(df[['PatientID','Date']].drop_duplicates(), how='inner', on=['PatientID','Date'])
    
    # Evaluation/Loss Inclusion Criteria wrt start/end date
    # 1. Not positive prior to end date.
    # 2. Not a prior case of CDI. 
    # 3. LOS of 7 prior to end date. (Note: #1 & #3 combined excludes community associated cases.)
    # 4. Not in the prior dataset
    # 5. Not a location node
    
    if last_day == None:
        df['evalloss_pop'] = 1*((df.nic_Cdiff==0))
    else:
        df['evalloss_pop'] = 1*((df.nic_Cdiff==0) & (df.AdmitDate>last_day))
                
    # Label: Positive, Collection after end_date, Admission overlaps start_date/end_date
    # Note: we can simply use Cdiff b/c of the application of the Network and Evaluation/loss inclusion criteria.
    criteria = df[['PatientID','evalloss_pop','Cdiff','nic_Cdiff','CollectionDateCdiff']].groupby('PatientID', as_index=False).agg(
        {'Cdiff':'max','evalloss_pop':'min','nic_Cdiff':'max','CollectionDateCdiff':'min'})
    # Note: we have to take a groupby instead of a duplicates drop for the scenario were two visits from the same patient occur between start & end. 
    
    
    ## KEY: NODE ID + LABELS ##
    key = loc.loc[:,['FacilityCode','Unit']].drop_duplicates()
    key['location'] = np.ones((key.shape[0]))
    print('--Num Locations: {}'.format(key.shape[0]))
    key = key.append(loc.PatientID.drop_duplicates().to_frame('PatientID'), sort=False)
    key['location'].fillna(0,inplace=True)   
    
    key = key.merge(criteria, how='left', on='PatientID')
    for col in criteria.columns:
        if (col=='PatientID') | (col=='CollectionDateCdiff'): continue
        key[col].fillna(0,inplace=True)
    print('--Num Patients: {:.0f}, Positive {}'.format(key.shape[0]-key.location.sum(),key.Cdiff.sum()))
    print('--Num Patients Post Exclusion: {:.0f}, Positive {}'.format(key.evalloss_pop.sum(),\
                                                       key.loc[key.evalloss_pop==1,'Cdiff'].sum()))
    
    key.reset_index(inplace=True, drop=True)
    key['id'] = key.index
    key['n'] = [key.shape[0]]*key.shape[0]
    key['T'] = [args['T']]*key.shape[0]
    args['n'] = key.shape[0]

    
    ## FEATURES & NETWORK ##
    # Features 
    feat_mat = np.zeros((args['T']*args['n'],4)) #location + timestep + nic_Cdiff(known case) + day of CDI pos

    # Generate Daily Graphs
    cG_list = []
    for timestep, day in enumerate(np.sort(loc.Date.unique())):

        # Feature Matrix        
        zfeat_df = key.copy()
        zfeat_df['Date'] = day
        
        # Cdipos Feature
        # -- if I just used a binary indicator for cdipos day, it would likely not fire b/c of the date range.
        zfeat_df['cdipos'] = (zfeat_df.CollectionDateCdiff-zfeat_df.Date).dt.days
        # -- for days without CollectionDateCdiff, cdipos is NaN
        zfeat_df.loc[zfeat_df.CollectionDateCdiff>=args['end_date'],'cdipos'] = np.nan
        # -- Cap and Normalize.
        zfeat_df['cdipos'] = np.maximum(zfeat_df.cdipos,-30)
        zfeat_df['cdipos'] = (zfeat_df['cdipos']+30)/30
        # -- cdipo = 1 => became positive that day
        # -- there are some days where cdipos>1 => this can happen if the pos day lies within date range. 
        # -- now when we fillna(0) - this makes sense. 
        
        # Feature Matrix
        zfeat_df.fillna(0, inplace=True)
        zfeat_df = zfeat_df.groupby('id')['location','nic_Cdiff','cdipos'].max()
        zfeat_df['timestep'] = [timestep/args['T']]*zfeat_df.shape[0]
                
        feat_mat[timestep*args['n']:timestep*args['n']+args['n'],:] = \
            zfeat_df.loc[:,['location','timestep','cdipos','nic_Cdiff']].values
        
        # Generate Graph
        adj_df = loc.loc[loc.Date==day,:]
        adj_df = adj_df.merge(key.loc[:,['PatientID','id']],how='left',on='PatientID')
        adj_df = adj_df.merge(key.loc[:,['FacilityCode','Unit','id']],how='left',\
                        on=['FacilityCode','Unit'])
        adj_df = adj_df.loc[:,['id_x','id_y']].drop_duplicates()
        adj_mat = np.zeros((args['n'],args['n']))
        adj_mat[adj_df.id_x.values,adj_df.id_y.values]=1
        adj_mat[adj_df.id_y.values,adj_df.id_x.values]=1
        cG = nx.from_numpy_matrix(adj_mat)
        cG_list.append(cG)
        
        # Assign Location & Timestep as Attributes
        attrs = {(i):{'time':timestep, \
                      'location':key.loc[key.id==i,'location'].values[0]} for i in range(args['n'])}
        nx.set_node_attributes(cG, attrs)

    #Rename nodes so that node names don't overlap when graphs combined
    #Nodes are numbered so that if n=10, node 0, 10, 20, 30, etc, refer to the same node but at t=0,1,2,3, etc
    for ind, graph in enumerate(cG_list):
        cG_list[ind] = nx.convert_node_labels_to_integers(graph, first_label=ind*(args['n']))
        #-> Be careful of the covert_node_labels_to_integers function (in our case it works out - but sometimes the sorted order is not the order you think it is)

    #Combine Daily Graphs
    cG = nx.compose_all(cG_list)
    
    #Create edges between past and future self. 
    for n in range(args['n']):
        for t in range(1,args['T']):
            cG.add_edge((t-1)*args['n'] + n ,t*args['n'] + n)
                    
    return cG, key, feat_mat
```

```python
# Generate and Save Datsets
args['n_list'] = []
args['T_list'] = []
key = pd.DataFrame({})
X = []
class_imbalance = []

count = 0 
for dataset in ['dates_train','dates_val','dates_test']:
    for start, end in args[dataset].copy():

        args['start_date'], args['end_date'] = start, end 
        cG, zkey, zX = df_to_network(loc, df, args)

        # If no positives - skip
        if zkey.loc[zkey.evalloss_pop==1,'Cdiff'].sum()==0:
            args[dataset].remove((start,end))
            continue
        
        # Save Stats
        args['n_list'].append(args['n'])
        args['T_list'].append(args['T'])    
        class_imbalance.append(np.mean(zkey.loc[zkey.evalloss_pop==1,'Cdiff'].values))
        
        # Save Data
        nx.write_gpickle(cG, os.path.join(args['savepath'],args['dataname'],'g'+str(count)+'.pkl'))
        zkey['dataset'] = [count]*zkey.shape[0]
        key = key.append(zkey, sort=False, ignore_index=True)
        X.append(zX)
        
        count += 1

# Calc Args (based on final dataset)
args['num_datasets_train'] = len(args['dates_train'])
args['num_datasets_val'] = len(args['dates_val'])
args['num_datasets_test'] = len(args['dates_test']) 
args['num_datasets'] = args['num_datasets_train'] + args['num_datasets_val'] + args['num_datasets_test']
args['dates'] = args['dates_train']+args['dates_val']+args['dates_test']

# Add Train/Val Class Imbalance Into ARgs
args['trainval_class_imb'] = np.mean(class_imbalance[:-args['num_datasets_test']])

# Save
key['global_id'] = np.arange(key.shape[0])
np.savez(os.path.join(args['savepath'],args['dataname'],'X.npz'),X)
key.to_hdf(os.path.join(args['savepath'],args['dataname'],'key.npz'),'key')
np.save(os.path.join(args['savepath'],args['dataname'],"args.npy"),args)

print('Class Imbalance: {:.3f}'.format(np.mean(class_imbalance)))
print('Class Imbalance Test: {:.3f}'.format(np.mean(class_imbalance[-args['num_datasets_test']:])))
```

### Calculate Bootstrapped AUC for best model

```python
from sklearn.utils import resample
def emp_bootstrap_auc(y_true, y_pred, sample_auroc, n=1000, percentile=0.95):
    
    scores = []
    while True:
        try:
            _y_true, _y_pred = resample(y_true, y_pred)
            scores.append(roc_auc_score(_y_true, _y_pred))
        except:
            pass
        if len(scores)>=n:
            break
        
    deltas = scores - sample_auroc
    deltas.sort()
    ub_idx = int(n*((1-percentile)/2)) # "p.5"
    lb_idx = int(n*(percentile+(1-percentile)/2)) # "p.95" 
    return(sample_auroc - deltas[lb_idx], sample_auroc - deltas[ub_idx])

def bootstrap_auc_from_model(runpath, dbundle):
    
    # Load args
    args = np.load(runpath+'args.npy',allow_pickle=True).item()
    
    # Load Model
    model = args['GCN'](args).cuda()
    model.load_state_dict(torch.load(runpath+'.pth.tar'))
    model.eval()

    #Evaluate
    yhat_y = {'ytest':[], 'yhattest':[]}
    # -- Test
    for batch_idx, (S, target, A, kA, pop, gid) in enumerate(dbundle.test_loader):
        _, yhat_, y_= auroc(args, model, S, target, A, kA, pop, gid, return_yhaty=True)
        yhat_y['ytest'].append(y_)
        yhat_y['yhattest'].append(yhat_)

    for k in yhat_y.keys():
        yhat_y[k] = np.concatenate(yhat_y[k], axis=0)
    yhat_y['auroctest'] = roc_auc_score(yhat_y['ytest'], yhat_y['yhattest'])

    # Calculate empirical bootstrap of the auc
    return yhat_y['auroctest'], emp_bootstrap_auc(yhat_y['ytest'], yhat_y['yhattest'], yhat_y['auroctest'])
```

```python
zdf = df.loc[df.groupby(["Model"])["best_val_auc"].idxmax()]
zdf['CI'] = ['']*zdf.shape[0]
for name in zdf.name:
    runpath = args['savepath']+'/results/'+name
    a, (b,c) = bootstrap_auc_from_model(runpath, data_bundle)
    zdf.loc[zdf.name==name,"CI"] = '{:.3f} ({:.3f},{:.3f})'.format(a,b,c)
```

### Calculate Bootstrapped AUC for best val auc

```python
def bootstrap_auc_from_model(runpath, dbundle, val=False):
    
    # Load args
    args = np.load(runpath+'args.npy',allow_pickle=True).item()
    
    # Load Model
    model = args['GCN'](args).cuda()
    model.load_state_dict(torch.load(runpath+'.pth.tar'))
    model.eval()

    # Val vs Test
    if val:
        loader = dbundle.val_loader
    else:
        loader = dbundle.test_loader
        
    #Evaluate
    yhat_y = {'ytest':[], 'yhattest':[]}
    # -- Test
    for batch_idx, (S, target, A, kA, pop, gid) in enumerate(loader):
        _, yhat_, y_= auroc(args, model, S, target, A, kA, pop, gid, return_yhaty=True)
        yhat_y['ytest'].append(y_)
        yhat_y['yhattest'].append(yhat_)

    for k in yhat_y.keys():
        yhat_y[k] = np.concatenate(yhat_y[k], axis=0)
    yhat_y['auroctest'] = roc_auc_score(yhat_y['ytest'], yhat_y['yhattest'])

    # Calculate empirical bootstrap of the auc
    return yhat_y['auroctest'], emp_bootstrap_auc(yhat_y['ytest'], yhat_y['yhattest'], yhat_y['auroctest'])
```
